```r
#must substring the data, so need to read_file first
data_str = read_file('../Charter-school-identities/data/wem_model_train250_nostem_unlapped_300d.txt')
```

```r
data = read_delim(substr(data_str, start=12, nchar(data_str)),delim=' ',col_names = F)
```

```r
#data cleaning
data = data.frame(data)
preserve_col = colnames(data)[-(ncol(data))]
row_n = t(data[1])
df = data[-1]
row.names(df)  = row_n
colnames(df) = preserve_col
head(df)
```

```r
nbclust = NbClust(df, distance='manhattan', min.nc=10, max.nc=30,
                  method='kmeans', index='all')
```

```r
#subsetting the cluster assignments from the output
cluster_group = nbclust.Best.partition
```

```r
pca <- prcomp(df, scale=T)
summary(pca)
```

```r
k2 = kmeans(df, centers=2, nstart=10)
k3 = kmeans(df, centers=3, nstart=10)
k4 = kmeans(df, centers=4, nstart=10)
```

```r
set.seed(12345)
df_sampled = sample_n(df, 4000)
head(df_sampled)
```

```r
nbclust = NbClust(df_sampled, distance='manhattan', min.nc=2, max.nc=30,
                  method='ward.D', index='all')
```
