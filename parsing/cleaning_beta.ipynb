# Cleaning charter data with pandas

```python
import pandas as pd
import numpy as np
```

```python
#read target file to data.
#input: filename(str), encoding_method(either "latin1" or "utf-8" depending on the language)
data = pd.read_csv("../pubschools_merged_2015.csv", encoding = "latin1", low_memory=False)
```

```python
#showing the length of data
print("Original length of data is " + str(len(data)))
```

```python
#filtering out schools that are considered as charter schools on 2015 or 2016.
#Can do other filtering by changing the conditions and using or.
data_charter = data[(data.CHARTER_TEXT16=='Yes')| (data.CHARTER_TEXT15=='Yes') | (data.CHARTER_TEXT == 'YES')]
print("Length of data after clearing non-charters is " + str(len(data_charter)))
```

```python
#filetering out schools that are opening at 2015 or 2016.
data_open_charter = data_charter[(data.SY_STATUS16==1)|(data.SY_STATUS16==3)|(data.SY_STATUS16==4)|\
                  (data.SY_STATUS16==5)|(data.SY_STATUS16==8)|(data.SY_STATUS15==1)|\
                  (data.SY_STATUS15==3)|(data.SY_STATUS15==4)|(data.SY_STATUS15==5)|\
                  (data.SY_STATUS15==8)|(data.SY_STATUS==1)|\
                  (data.SY_STATUS==3)|(data.SY_STATUS==4)|(data.SY_STATUS==5)|\
                  (data.SY_STATUS==8)]
print("Length of data after clearing closed charters is " + str(len(data_open_charter)))
```

```python
#Showing the number of rows in our data after the cleaning
data_open_charter2 = data_charter[(np.isnan(data.YEAR_CLOSED))| (data.YEAR_CLOSED > 2014)]
print("Length of data after clearing closed charters is " + str(len(data_open_charter2)))
```
