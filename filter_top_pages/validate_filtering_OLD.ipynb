```python
def count_pages(text):
    counter = 0 #count the number of urls 
    http_array = []
    for i in text:
        if i=='h':
            http_array.append("h")
        if i == 't' and len(http_array)== 1 and http_array[0] == 'h':
            http_array.append("t")
        if i == 't' and len(http_array)== 2 and http_array[1] == 't':
            http_array.append("t") 
        if i == 'p' and len(http_array)== 3 and http_array[2] == 't':
            print(http_array)
            counter += 1
            http_array = []
    return counter 
            
            
```

```python
def count_urls(text):
    urls = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', text)
    return len(urls)
```
